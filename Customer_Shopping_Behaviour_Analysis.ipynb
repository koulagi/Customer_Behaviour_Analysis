```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
df = df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
df.columns
```

```python
#Create a column age group
labels = ['Young Adult', 'Adult', 'Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
df[['age','age_group']].head(10)
```

```python
#create column purchase_frequency-days
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly':14,
    'Annually' : 365,
    'Every 3 Months' : 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
import pandas as pd
from sqlalchemy import create_engine
import urllib

# 1. Provide your SQL Server information
# (Copy the server name directly from the top of your SSMS Object Explorer)
server_name = r'SSKoulagi\SQLEXPRESS'  
database_name = 'CustomerBehaviour'

# FIXED: Swapped out {SQL Server} for {ODBC Driver 17 for SQL Server}
params = urllib.parse.quote_plus(
    f"DRIVER={{ODBC Driver 17 for SQL Server}};"
    f"SERVER={server_name};" 
    f"DATABASE={database_name};"
    "Trusted_Connection=yes;"
)

engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")
print("🎉 Connection Engine Updated!")

```

```python
# Upload your 3,900 rows straight to the database
df.to_sql(
    name='customer',  # The name you want for your new table in SSMS
    con=engine,                   # Your active connection engine
    schema='behave',           # ADDED: This forces it into your custom schema
    if_exists='replace',          # Overwrites the table if it already exists
    index=False                   # Prevents row numbers from turning into a column
)

print("🚀 3,900 rows successfully exported to SQL Server!")
```
